```python
def deep_merge_dict(base, override):
    merged = deepcopy(base)
    for key, value in (override or {}).items():
        if isinstance(value, dict) and isinstance(merged.get(key), dict):
            merged[key] = deep_merge_dict(merged[key], value)
        else:
            merged[key] = deepcopy(value)
    return merged


def build_headers(api_item):
    headers = {}
    auth_type = (api_item or {}).get("auth_type", "token")
    if auth_type in {"token", "token+bearer", "both"}:
        headers["token"] = TOKEN
    if auth_type in {"bearer", "token+bearer", "both"}:
        headers["authorization"] = f"Bearer {TOKEN}"
    headers.update((api_item or {}).get("headers", {}))
    return headers


def build_url(base_url, path):
    return f"{(base_url or DEFAULT_BASE_URL).rstrip('/')}/{path.lstrip('/')}"


def list_apis():
    rows = []
    for api_name, api_item in API_CONFIG.items():
        rows.append(
            {
                "api_name": api_name,
                "title": api_item.get("title"),
                "endpoint": api_item.get("endpoint"),
                "auth_type": api_item.get("auth_type"),
                "base_url": api_item.get("base_url"),
                "path": api_item.get("path"),
            }
        )
    df = pd.DataFrame(rows)
    display(df)
    return df


def post_api(api_name, payload_override=None, timeout=30):
    if api_name not in API_CONFIG:
        raise KeyError(f"Unknown api_name: {api_name}")
    api_item = API_CONFIG[api_name]
    payload = deepcopy(api_item.get("payload", {}))
    if payload_override:
        payload = deep_merge_dict(payload, payload_override) if isinstance(payload_override, dict) else payload_override
    url = build_url(api_item.get("base_url", DEFAULT_BASE_URL), api_item.get("path", ""))
    method = api_item.get("method", "POST").upper()
    headers = build_headers(api_item)
    request_kwargs = {"headers": headers, "timeout": timeout}
    if method == "GET":
        request_kwargs["params"] = payload
    else:
        request_kwargs["json"] = payload
    response = session.request(method, url, **request_kwargs)
    text = response.text
    try:
        body = response.json()
    except ValueError:
        body = None
    return {
        "api_name": api_name,
        "title": api_item.get("title"),
        "url": url,
        "status_code": response.status_code,
        "ok": response.ok,
        "json": body,
        "text": text,
        "payload": payload,
    }


def run_one(api_name, show_json=False):
    result = post_api(api_name)
    summary = pd.DataFrame(
        [
            {
                "api_name": result["api_name"],
                "title": result.get("title"),
                "status_code": result["status_code"],
                "ok": result["ok"],
                "url": result["url"],
            }
        ]
    )
    display(summary)
    if show_json:
        if result["json"] is not None:
            print(json.dumps(result["json"], ensure_ascii=False, indent=2))
        else:
            print(result["text"])
    return result


def run_all(api_names=None):
    api_names = list(api_names or API_CONFIG.keys())
    results = {}
    summary_rows = []
    for api_name in api_names:
        result = post_api(api_name)
        results[api_name] = result
        summary_rows.append(
            {
                "api_name": api_name,
                "title": result.get("title"),
                "status_code": result["status_code"],
                "ok": result["ok"],
                "url": result["url"],
            }
        )
    summary_df = pd.DataFrame(summary_rows)
    display(summary_df)
    return {"results": results, "summary": summary_df}


def _obj_to_df(obj):
    if isinstance(obj, list):
        return pd.DataFrame(obj)
    if isinstance(obj, dict):
        for value in obj.values():
            if isinstance(value, list):
                return pd.DataFrame(value)
        return pd.json_normalize(obj)
    return None


def to_df(result, data_key="data"):
    payload = result.get("json") if isinstance(result, dict) and "json" in result else result
    if payload is None:
        return None
    if isinstance(payload, list):
        return pd.DataFrame(payload)
    if not isinstance(payload, dict):
        return None
    for key in [data_key, "data", "retdata", "Data"]:
        if key in payload:
            df = _obj_to_df(payload[key])
            if df is not None:
                return df
    for value in payload.values():
        if isinstance(value, list):
            return pd.DataFrame(value)
    return None


def normalize_date(date_value):
    if isinstance(date_value, pd.Timestamp):
        return date_value
    return pd.to_datetime(str(date_value))


def shift_date(date_value, years=0, months=0):
    ts = normalize_date(date_value)
    return ts + pd.DateOffset(years=years, months=months)


def detect_date_style(field_name, current_value):
    compact_fields = {"bdate", "edate", "salebdate", "saleedate", "birthbdate", "birthedate"}
    if isinstance(current_value, str) and current_value:
        if "-" in current_value:
            return "dash"
        if current_value.isdigit() and len(current_value) == 8:
            return "compact"
    if field_name in compact_fields:
        return "compact"
    return "dash"


def format_date(date_value, style):
    ts = normalize_date(date_value)
    return ts.strftime("%Y%m%d" if style == "compact" else "%Y-%m-%d")


def _update_dates_in_obj(obj, start_date, end_date):
    start_keys = {"bdate", "BeginDate", "beginDate", "salebdate", "StartDate"}
    end_keys = {"edate", "EndDate", "endDate", "saleedate"}
    shifted_fields = {
        "YBeginDate": (start_date, -1, 0),
        "YEndDate": (end_date, -1, 0),
        "MBeginDate": (start_date, 0, -1),
        "MEndDate": (end_date, 0, -1),
    }
    if isinstance(obj, dict):
        for key, value in obj.items():
            if isinstance(value, (dict, list)):
                _update_dates_in_obj(value, start_date, end_date)
            elif key in start_keys:
                obj[key] = format_date(start_date, detect_date_style(key, value))
            elif key in end_keys:
                obj[key] = format_date(end_date, detect_date_style(key, value))
            elif key in shifted_fields:
                ref_date, years, months = shifted_fields[key]
                obj[key] = format_date(
                    shift_date(ref_date, years=years, months=months),
                    detect_date_style(key, value),
                )
    elif isinstance(obj, list):
        for item in obj:
            _update_dates_in_obj(item, start_date, end_date)


def update_date_range(start_date, end_date, api_names=None):
    api_names = list(api_names or API_CONFIG.keys())
    for api_name in api_names:
        _update_dates_in_obj(API_CONFIG[api_name].get("payload"), start_date, end_date)
    return {api_name: API_CONFIG[api_name]["payload"] for api_name in api_names}


def _update_fields_in_obj(obj, replacements):
    if isinstance(obj, dict):
        for key, value in obj.items():
            if key in replacements:
                obj[key] = replacements[key]
            elif isinstance(value, (dict, list)):
                _update_fields_in_obj(value, replacements)
    elif isinstance(obj, list):
        for item in obj:
            _update_fields_in_obj(item, replacements)


def update_common_filters(api_names=None, **filters):
    api_names = list(api_names or API_CONFIG.keys())
    alias_map = {
        "depart": ["Depart", "depts"],
        "depts": ["depts"],
        "spenum": ["spenum"],
        "warecause": ["warecause", "WareClause"],
        "name": ["name"],
        "search": ["Search", "searchval"],
        "searchval": ["searchval"],
        "condition": ["condition"],
        "operater": ["Operater"],
        "tiem": ["Tiem"],
        "datetype": ["datetype"],
        "doctype": ["doctype"],
        "stockflag": ["stockflag"],
        "tag": ["tag"],
        "menuid": ["menuid", "MenuID"],
        "gridid": ["gridid"],
        "search_type": ["SearchType"],
        "type": ["type", "Type"],
    }
    replacements = {}
    for key, value in filters.items():
        for alias in alias_map.get(key, [key]):
            replacements[alias] = value
    for api_name in api_names:
        _update_fields_in_obj(API_CONFIG[api_name].get("payload"), replacements)
    return {api_name: API_CONFIG[api_name]["payload"] for api_name in api_names}
```
